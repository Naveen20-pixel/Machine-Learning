## Linear Regression Model

```python
## Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Importing Dataset
df=pd.read_csv('house.csv')
```

```python
## Creating Dependent and Independent Variables
x=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
## Splitting Dataset into Train and Test Set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
## Building Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn import metrics
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print("Accuracy of the Model:",score*100)
```

```python
## Visualizing the Model
plt.scatter(x,y,color='green')
plt.plot(x_train,reg.predict(x_train),color='orange')
plt.title('Linear Regression Model',size=30,color='red')
plt.xlabel('Area',size=20,color='blue')
plt.ylabel('Priza',size=20,color='blue')
plt.show()
```
